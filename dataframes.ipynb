```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter("ignore", FutureWarning)
```

```python
dataset = "News.json"

if os.path.isfile(dataset):
    news = pd.read_json("News.json", lines=True)

display(news.head(10))
```

```python
headline = pd.read_csv('headline_wv.csv', index_col=0)
headline.head()
```

```python
#short_description = pd.read_csv('short_wv.csv', index_col=0)
#short_description.head()
```

```python
#THE WORLDPOST and WORLDPOST should be the same category, so merge them. (and others as well)
def category_cleaner(x):
    
    if x == 'THE WORLDPOST':
        return 'WORLDPOST'
    elif x == 'PARENTING':
        return 'PARENTS'
    elif x == 'ARTS' or x == 'CULTURE & ARTS':
        return 'ARTS & CULTURE'
    elif x == 'STYLE':
        return 'STYLE & BEAUTY'
    elif x == 'COLLEGE':
        return 'EDUCATION'
    elif x == 'TASTE':
        return 'FOOD & DRINK'
    else:
        return x
    
news['category'] = news.category.apply(category_cleaner)
```

```python
headline['category'] = news.category
headline.head()
```

```python
headline.columns
```

```python
features = headline.columns[:300] 
target = 'category'
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(headline[features], headline['category'], test_size=0.2, random_state=42)
X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score

logreg = LogisticRegression(random_state=0, class_weight='balanced', C=1.0)
logreg.fit(X_train, y_train)
preds = logreg.predict(X_test)


print("micro:", f1_score(y_test, preds, average='micro'))
print("I am done!")
```
